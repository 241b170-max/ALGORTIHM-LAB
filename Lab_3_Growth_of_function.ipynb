# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

```python
import matplotlib.pyplot as plt

def check_even_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

n_values = list(range(1, 101))
time_values = []


for n in n_values:
    check_even_odd(n)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values, marker='o', linestyle='--', color='purple')
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Even/Odd Check")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
import time
import matplotlib.pyplot as plt

def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 5000, 100))  # larger range for clearer scaling
time_values = []

for n in n_values:
    data = list(range(n))

    start = time.time()
    linear_time_sum(data)
    end = time.time()

    elapsed = end - start
    time_values.append(elapsed)

plt.figure(figsize=(8,5))
plt.plot(n_values, time_values, marker='o')
plt.xlabel("Input size n")
plt.ylabel("Execution time (seconds)")
plt.title("O(n) Linear Time Complexity")
plt.grid(True)
plt.show()
```

```python
import time
import matplotlib.pyplot as plt

def even_odd_counts(numbers):
    even_count = 0
    odd_count = 0
    for num in numbers:
        if num > 0:
            if num % 2 == 0:
                even_count += 1
            else:
                odd_count += 1
    return even_count, odd_count

n_values = list(range(100, 5000, 200))
time_values = []

for n in n_values:
    data = list(range(-n, n))
    start = time.time()
    even_odd_counts(data)
    end = time.time()

    elapsed = end - start
    time_values.append(elapsed)

plt.figure(figsize=(8,5))
plt.plot(n_values, time_values, marker='o', color='purple')
plt.xlabel("Input size n")
plt.ylabel("Execution time (seconds)")
plt.title("O(n) Linear Time Complexity: Even/Odd Counts > 0")
plt.grid(True)
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
```

```python
import time
import matplotlib.pyplot as plt

def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

n_values = list(range(1000, 50000, 2000))
time_values = []

for n in n_values:
    data = list(range(n))  # sorted list
    target = n - 1         # search for last element

    start = time.time()
    binary_search(data, target)
    end = time.time()

    elapsed = end - start
    time_values.append(elapsed)

plt.figure(figsize=(8,5))
plt.plot(n_values, time_values, marker='o', color='green')
plt.xlabel("Input size n")
plt.ylabel("Execution time (seconds)")
plt.title("O(log n) Logarithmic Time Complexity: Binary Search")
plt.grid(True)
plt.show()


```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.
```

```python
import time
import random
import matplotlib.pyplot as plt

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

n_values = list(range(1000, 20000, 2000))
time_values = []

for n in n_values:
    data = [random.randint(0, n) for _ in range(n)]  # random list

    start = time.time()
    merge_sort(data)
    end = time.time()

    elapsed = end - start
    time_values.append(elapsed)


plt.figure(figsize=(8,5))
plt.plot(n_values, time_values, marker='o', color='red')
plt.xlabel("Input size n")
plt.ylabel("Execution time (seconds)")
plt.title("O(n log n) Log-Linear Time Complexity: Merge Sort")
plt.grid(True)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
```

```python
import time
import random
import matplotlib.pyplot as plt

# Example function: Sorting (O(n log n))
def sort_numbers(arr):
    return sorted(arr)   # Timsort, average O(n log n)

# Input sizes
n_values = list(range(1000, 20000, 2000))
time_values = []

for n in n_values:
    data = [random.randint(0, n) for _ in range(n)]  # random list

    start = time.time()
    sort_numbers(data)
    end = time.time()

    elapsed = end - start
    time_values.append(elapsed)

# Plotting the runtime
plt.figure(figsize=(8,5))
plt.plot(n_values, time_values, marker='o', color='orange')
plt.xlabel("Input size n")
plt.ylabel("Execution time (seconds)")
plt.title("O(n log n) Log-Linear Time Complexity: Sorting")
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the input size range
n_values = np.arange(1, 101)

# Define time complexity functions
def o_1(n): # Constant Time
    return np.ones_like(n)

def o_log_n(n): # Logarithmic Time
    return np.log(n)

def o_n(n): # Linear Time
    return n

def o_n_log_n(n): # Log-Linear Time
    return n * np.log(n)

def o_n_squared(n): # Polynomial Time (n^2)
    return n**2

def o_2_power_n(n): # Exponential Time (2^n)
    return 2**n

# Calculate time values for each complexity
time_o_1 = o_1(n_values)
time_o_log_n = o_log_n(n_values)
time_o_n = o_n(n_values)
time_o_n_log_n = o_n_log_n(n_values)
time_o_n_squared = o_n_squared(n_values)
time_o_2_power_n = o_2_power_n(n_values)

# Plotting
plt.figure(figsize=(12, 8))

plt.plot(n_values, time_o_1, label='O(1) - Constant', color='purple')
plt.plot(n_values, time_o_log_n, label='O(log n) - Logarithmic', color='green')
plt.plot(n_values, time_o_n, label='O(n) - Linear', color='blue')
plt.plot(n_values, time_o_n_log_n, label='O(n log n) - Log-Linear', color='cyan')
plt.plot(n_values, time_o_n_squared, label='O(n^2) - Polynomial', color='orange')
plt.plot(n_values, time_o_2_power_n, label='O(2^n) - Exponential', color='red')

plt.xlabel('Input Size (n)')
plt.ylabel('Operations / Time')
plt.title('Common Time Complexities (Growth of Functions)')
plt.ylim(0, 100) # Limit y-axis for better visibility of lower complexities
plt.xlim(0, 100) # Limit x-axis to match input size range
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Write a function to perform quick sort and show that it has Logarithmic Time Complexity O(log n)
```

```python
import time
import random
import matplotlib.pyplot as plt

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Input sizes
n_values = list(range(1000, 20000, 2000))
time_values = []

for n in n_values:
    data = [random.randint(0, n) for _ in range(n)]  # random list

    start = time.time()
    quick_sort(data)
    end = time.time()

    elapsed = end - start
    time_values.append(elapsed)

# Plotting
plt.figure(figsize=(8,5))
plt.plot(n_values, time_values, marker='o', color='blue')
plt.xlabel("Input size n")
plt.ylabel("Execution time (seconds)")
plt.title("Quick Sort Average Time Complexity: O(n log n)")
plt.grid(True)
plt.show()
```
